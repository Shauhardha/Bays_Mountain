## Encounter Type: Medication

```python
medicationLog_df = medLog_df[medLog_df['Encounter Type'] == "Medication"]

medicationCols = ['Timestamp','Animal','Staff','Date','Time','Animal Name','Type.1','Dose.1','Administration Route','Did animal take medication?']

medicationLog = medicationLog_df[medicationCols]

print(medicationLog.head())

print(medicationLog.shape)
```

```python
import numpy as np

np.sort(medicationLog['Type.1'].unique())
```

```python
medicationLog[['Type.1','Dose.1','Administration Route','Did animal take medication?']].nunique()
```

```python
medical_df = medicationLog.copy()

keywords = {'Meloxicam','Cephalexin','Gabapentin','Bravecto','Intercepter'}

# Function to process the column
def split_conditions(description):
    # Split the string into parts
    items = [item.strip() for item in description.split(',')]
    
    # Identify matching and non-matching items
    matched = {kw: kw in items for kw in keywords}
    remaining = ', '.join([item for item in items if item not in keywords])
    
    return pd.Series({**matched, 'Other': remaining if remaining else None})

# Apply the function
medical_df = medical_df.join(medical_df['Type.1'].apply(split_conditions))

# Display the result
print(medical_df)
```

### Medication Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = medical_df.copy()

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
#Barchart to count each medication type

top_medication = df[['Meloxicam','Cephalexin','Gabapentin','Bravecto','Intercepter']].sum()
top_medication.plot(kind='bar', figsize=(8,5), title='Medication Type Count')
plt.xlabel("Medication Type")
plt.ylabel("Count")
plt.show()
```

```python
#Number of Medication given per Animal

df['Animal'].value_counts().plot(kind='bar', figsize=(8,5), title='Medication giver per Animal')
plt.xlabel("Animal")
plt.ylabel("Count")
plt.show()
```

```python
#Medication administered by each Staff member

df['Staff'].value_counts().plot(kind='bar', figsize=(8,5), title="Medication Administered by each Staff")
plt.xlabel('Staff')
plt.ylabel('Count')
plt.show()
```

```python
#Pie Chart: Success rate of medication administration

df['Did animal take medication?'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Medication Success Rate')
plt.ylabel("")
plt.show()
```

```python
# Grouped Bar Chart: Medication Type vs. Animal Type
medication_counts = df.groupby('Animal')[['Gabapentin', 'Bravecto', 'Meloxicam', 'Intercepter', 'Cephalexin']].sum()
medication_counts.plot(kind='bar', stacked=True, figsize=(10, 6), title='Medication Types per Animal')
plt.xlabel("Animal")
plt.ylabel("Count")
plt.legend(title='Medication Type')
plt.show()
```

```python
# Line Chart: Medications given over time
#df['Date'] = pd.to_datetime(df['Date'])

med_col = ['Meloxicam','Cephalexin','Gabapentin','Bravecto','Intercepter']

for med in med_col:
    time_series = df.groupby('Timestamp')[med].count()
    time_series.plot(kind='line', figsize=(10, 5), marker='o', title=f'{med} Administration Over Time')
    plt.xlabel("Date")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python

from IPython.display import display, Markdown

# Convert 'Timestamp' column to datetime if not already
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

med_col = ['Meloxicam', 'Cephalexin', 'Gabapentin', 'Bravecto', 'Intercepter']

for med in med_col:
    display(Markdown(f"### {med} Administration Over Time"))  # Show as markdown in Jupyter

    # Create figure and axis
    fig, ax = plt.subplots(figsize=(10, 5))

    # Plot time series data
    time_series = df.groupby(df['Timestamp'].dt.date)[med].count()
    time_series.plot(kind='line', marker='o', title=f'{med} Administration Over Time', ax=ax)

    ax.set_xlabel("Date")
    ax.set_ylabel("Count")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

    # Display the plot
    plt.show()
```

```python
# Heatmap: Frequency of medication combinations
med_combo = df[['Gabapentin', 'Bravecto', 'Meloxicam', 'Intercepter', 'Cephalexin']].astype(int)
plt.figure(figsize=(8, 6))
sns.heatmap(med_combo.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Medication Combination Heatmap")
plt.show()
```
